```python
from skimage import io
from skimage.metrics import structural_similarity
```

```python
img_cat1 = io.imread('cat1.jpg')
img_cat2 = io.imread('cat2.jpg')
img_dog = io.imread('dog1.jpg')
```

```python
import cv2
import numpy as np
```

```python
img_cat2 = cv2.resize(img_cat2, dsize=img_cat1.shape[:2])
img_dog = cv2.resize(img_dog, dsize=img_cat1.shape[:2])
```

```python
ssim_original = structural_similarity(img_cat1, img_cat1,
                    data_range=img_cat1.max() - img_cat1.min(), multichannel=True)
```

```python
ssim_original
```

```python
ssim_cat = structural_similarity(img_cat1, img_cat2,
               data_range=img_cat2.max()-img_cat2.min(), multichannel=True)
```

```python
ssim_cat
```

```python
ssim_dog = structural_similarity(img_cat1, img_dog,
               data_range=img_cat1.max()-img_cat1.min(), multichannel=True)
```

```python
ssim_dog
```
